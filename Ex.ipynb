```python
import pandas as pd
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df
```

```python
df.info()
```

```python
df.satisfaction_level = df.satisfaction_level.fillna(df.satisfaction_level.median())
df.last_evaluation = df.last_evaluation.fillna(df.last_evaluation.median())
```

```python
df2 = pd.get_dummies(df)
# df.salary = pd.get_dummies(df.salary)
```

```python
df2
```

```python
x = df2.drop(['left'], axis = 'columns')
x
```

```python
y = df2.left
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
lc = LogisticRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2, random_state = 79)
x_train
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2, random_state = 990)
```

```python
lc.fit(x_train,y_train)
```

```python
lc.score(x_test,y_test)
```
